```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv("Datos.csv")
```

```python
model = (smf.ols(formula= 'PUNT_GLOBAL ~ male + etnia_si + internet_si + computador_si + cole_academico + calendario_b + cole_bilingue + cole_urbano' ,data= df).fit())
model.params
print(model.summary())
```
