```python
# Sacar los promedios del puntaje global para las variables dummies
mean_internet = data.groupby(['internet_si']).PUNT_GLOBAL.mean()
mean_comp = data.groupby(['computador_si']).PUNT_GLOBAL.mean()
mean_etnia = data.groupby(['etnia_si']).PUNT_GLOBAL.mean()
mean_genero = data.groupby(['male']).PUNT_GLOBAL.mean()
mean_trabaja = data.groupby(['trabaja_no']).PUNT_GLOBAL.mean()

mean_zona = data.groupby(['cole_rural']).PUNT_GLOBAL.mean()
mean_cole = data.groupby(['cole_publico']).PUNT_GLOBAL.mean()
mean_cal = data.groupby(['calendario_b']).PUNT_GLOBAL.mean()
mean_biling = data.groupby(['cole_bilingue']).PUNT_GLOBAL.mean()
mean_academico = data.groupby(['cole_academico']).PUNT_GLOBAL.mean()

mean_econsit = data.groupby(['FAMI_SITUACIONECONOMICA']).PUNT_GLOBAL.mean()

```

```python
print(mean_trabaja)
```

```python
#Guardarlas en diccionarios
ind_variables = { "Tiene Internet": mean_internet, "Tiene Computador": mean_comp ,
             "Tiene Etnía": mean_etnia, "Es Masculino": mean_genero, "No Trabaja": mean_trabaja}

cole_variables = { "Es Rural": mean_zona, "Es Público": mean_cole ,
             "Es Calendario B": mean_cal, "Es Bilingue": mean_biling, "Es Académico" : mean_academico}

# convertirlo en un dataframe
ind_variables = pd.DataFrame.from_dict(ind_variables)
ind_variables
```

```python
cole_variables = pd.DataFrame.from_dict(cole_variables)
cole_variables
```

```python
#Transponer
ind_variables= ind_variables.T
```

```python
ind_variables.columns
```

```python
ind_variables['Tipo Desigualdad'] = ind_variables.index
ind_variables.columns.values[0] = "No"
ind_variables.columns.values[1] = "Yes"
ind_variables
```

```python
ind_variables.columns
```

```python
ind_variables = ind_variables.astype({"No":"int","Yes":"int"})
```

```python
# Crear un gráfico lollipop para comparar las diferencias en puntaje global para las variables dummies
fig, ax = plt.subplots(figsize=(8,8))

#Crear los puntos mínimos y máximos
ordered_df = ind_variables.sort_values(by ='No')
my_range=range(1,len(ind_variables.index)+1)
ax.hlines(y=my_range, xmin=ordered_df['No'], xmax=ordered_df['Yes'], color='grey', alpha=0.4)
ax.scatter(ordered_df['No'], my_range, color='skyblue' ,alpha=1, label='No')
ax.scatter(ordered_df['Yes'], my_range, color='green', alpha=0.5 , label='Sí')
ax.legend()

# Title and axis names
plt.yticks(my_range, ordered_df['Tipo Desigualdad'])
plt.title("Factores Familiares de la Desigualdad", loc='left')
plt.xlabel('Puntaje Global Promedio')
plt.ylabel('Variable de Desigualdad')


# Agragar data points
for i, txt in enumerate(ordered_df['Yes']):
    plt.annotate(txt, (ordered_df['Yes'][i], my_range[i]), 
                 xytext=(0,5), textcoords='offset pixels')
for i, txt in enumerate(ordered_df['No']):
    plt.annotate(txt, (ordered_df['No'][i], my_range[i]), 
                 xytext=(0, 5), textcoords='offset pixels')    

plt.savefig('factoresfamiliares.png')
plt.show()
```
