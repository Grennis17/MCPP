```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("Data_collapsed.csv")
```

## Nivel Escolar

```python
data.head()
```

```python
data = data.astype({"COLE_COD_DEPTO_UBICACION":"int","COLE_COD_MCPIO_UBICACION":"int"})
data.dtypes
```

```python
data = data.astype({"cole_publico":"int","cole_urbano":"int", "cole_privado":"int","cole_rural":"int", "PERCENTIL_GLOBAL":"int"})
```

```python
data.dtypes
```

```python
data['PERCENTIL_GLOBAL']
```

```python
data.head()
```

```python
pyramid_data = (data
                    .groupby(['PERCENTIL_GLOBAL' ])
                    .agg({'PERCENTIL_GLOBAL': 'mean',
                          'cole_publico' : 'sum',
                          'cole_privado' : 'sum',
                          'cole_rural' : 'sum',
                          'cole_urbano': 'sum',
                          }))
```

```python
pyramid_data.head()
```

```python
## Convertir cada categoría en porcentajes del total para que la escala es igual (ya que hay muchos más colegios públicos que privados)
pyramid_data['pub_perc'] = (pyramid_data['cole_publico'] / pyramid_data['cole_publico'].sum()) * 100
pyramid_data['pri_perc'] = (pyramid_data['cole_privado'] / pyramid_data['cole_privado'].sum()) * 100
pyramid_data['urb_perc'] = (pyramid_data['cole_urbano'] / pyramid_data['cole_urbano'].sum()) * 100
pyramid_data['rur_perc'] = (pyramid_data['cole_rural'] / pyramid_data['cole_rural'].sum()) * 100
pyramid_data
```

```python
#pyramid_data = pyramid_data.astype({"pub_perc":"int","pri_perc":"int", "urb_perc":"int","rur_perc":"int", "PERCENTIL_GLOBAL":"int"})
#pyramid_data
```

```python
import plotly.graph_objects as gp

puntaje = pyramid_data['PERCENTIL_GLOBAL']
x_publico = pyramid_data['pub_perc']
x_privado = pyramid_data['pri_perc'] * -1


# Creating instance of the figure
fig = gp.Figure()
  
# Adding cole_publico data to the figure
fig.add_trace(gp.Bar(y= puntaje, x = x_publico, 
                     name = 'Colegios Publicos', 
                     orientation = 'h'))
  
# Adding cole_privado data to the figure
fig.add_trace(gp.Bar(y = puntaje, x = x_privado,
                     name = 'Colegios Privados', orientation = 'h'))
  
# Updating the layout for our graph
fig.update_layout(title = 'Percentil de Puntaje Global según Estado del Colegio',
                 title_font_size = 18, barmode = 'relative',
                 bargap = 0.0, bargroupgap = 0,
                 xaxis = dict(tickvals = [-10, -8, -6, -4, -2,
                                          0, 2, 4, 6, 8, 10],
                                
                              ticktext = ['10%', '8%', '6%', '4%', '2%', '0', 
                                          '2%', '4%', '6%', '8%', '10%'],
                                
                              title = 'Porcentaje de Colegios',
                              title_font_size = 14)
                 )

fig.update_layout(
    yaxis_title="Percentil de Puntaje Global",
    legend_title="Colegios",
    font=dict(
        family="Arial",
        size=14,
        color="Black")
)

fig.show()
```

```python
puntaje = pyramid_data['PERCENTIL_GLOBAL']
x_urbano = pyramid_data['urb_perc']
x_rural = pyramid_data['rur_perc'] * -1


# Creating instance of the figure
fig = gp.Figure()
  
# Adding cole_publico data to the figure
fig.add_trace(gp.Bar(y= puntaje, x = x_urbano, 
                     name = 'Colegios Urbanos', 
                     orientation = 'h'))
  
# Adding cole_privado data to the figure
fig.add_trace(gp.Bar(y = puntaje, x = x_rural,
                     name = 'Colegios Rurales', orientation = 'h'))
  
# Updating the layout for our graph
fig.update_layout(title = 'Percentil de Puntaje Global según Estado del Colegio',
                 title_font_size = 18, barmode = 'relative',
                 bargap = 0.0, bargroupgap = 0,
                 xaxis = dict(tickvals = [-10, -8, -6, -4, -2,
                                          0, 2, 4, 6, 8, 10],
                                
                              ticktext = ['10%', '8%', '6%', '4%', '2%', '0', 
                                          '2%', '4%', '6%', '8%', '10%'],
                                
                              title = 'Porcentaje de Colegios',
                              title_font_size = 14)
                 )

fig.update_layout(
    yaxis_title="Percentil de Puntaje Global",
    legend_title="Colegios",
    font=dict(
        family="Arial",
        size=14,
        color="Black")
)

fig.show()
```

### Nivel Municipio

```python
data_collapsed_muni = (data
                    .groupby(['COLE_COD_DEPTO_UBICACION','COLE_COD_MCPIO_UBICACION'])
                    .agg({'COLE_COD_DEPTO_UBICACION': 'mean',
                          'COLE_COD_MCPIO_UBICACION' : 'mean',
                          'COLE_COD_DANE_ESTABLECIMIENTO' : 'sum',
                          'PUNT_LECTURA_CRITICA': 'mean', 'PERCENTIL_LECTURA_CRITICA': 'mean',
                          'PUNT_MATEMATICAS':'mean', 'PERCENTIL_MATEMATICAS': 'mean',
                          'PUNT_C_NATURALES' : 'mean','PERCENTIL_C_NATURALES': 'mean',
                          'PUNT_SOCIALES_CIUDADANAS': 'mean','PERCENTIL_SOCIALES_CIUDADANAS': 'mean',
                          'PUNT_INGLES': 'mean','PERCENTIL_INGLES': 'mean', 
                          'PUNT_GLOBAL': 'mean','PERCENTIL_GLOBAL': 'mean',
                          'cole_privado': 'sum',
                          'cole_publico': 'sum',
                          'cole_urbano': 'sum',
                          'cole_rural': 'sum',
                          'etnia_si': 'sum','internet_si': 'sum',
                          'computador_si': 'sum','ID' : 'sum'}))
```

```python
data_collapsed_muni.head()
```

```python
data = data_collapsed_muni
```

```python
data['etnia_perc'] =((data['etnia_si'] /data['ID'])*100)
data['int_perc'] =(data['internet_si'] /data['ID'])* 100
data['comp_perc'] =(data['computador_si'] /data['ID'])* 100
data['rural_share'] =(data['cole_rural'] /data['COLE_COD_DANE_ESTABLECIMIENTO']*100)
data['privado_share'] =(data['cole_privado'] /data['COLE_COD_DANE_ESTABLECIMIENTO']*100)
data.head()
```

```python
data = data.astype({"COLE_COD_DEPTO_UBICACION":"int","COLE_COD_MCPIO_UBICACION":"int"})
```

```python
data.head()
```

```python
data.privado_share.describe()
```

```python
data.shape
```

```python
plt.figure(figsize=(6,6)) # establecer el tamaño 
sns.scatterplot(data=data, x="int_perc", y="comp_perc", hue = 'privado_share').set(title='Conectividad del Municipio')
plt.xlabel('Porcentaje de Estudiantes con Conexión de Internet')
plt.ylabel('Porcentaje de Estudiantes que Tiene Computador')
plt.legend(title = '% Colegios Privados')
plt.savefig('muni_comp_int.png')
plt.show()
```

```python
plt.figure(figsize=(6,6)) # establecer el tamaño 
sns.scatterplot(data=data, x="PERCENTIL_GLOBAL", y="privado_share", hue = 'etnia_perc').set(title='Colegios Privados')
plt.xlabel('Percentil del Puntaje Global')
plt.ylabel('Porcentaje de Colegios Privados en Municipio')
plt.legend(title = '% Etnia')
plt.savefig('muni_pri_percentil_etnia.png')
plt.show()
```

### Departamento

```python
data2 = pd.read_csv("Deserscion.csv")
```

```python
df = pd.merge(left = data, right= data2, how= 'inner', left_on= "code_dept", right_on= "c_digo_departamento") 
df.head()
```

```python
departamento = (df.
                groupby(['code_dept'])
                    .agg({'code_dept' : 'mean',
                          'PUNT_GLOBAL': 'mean','PERCENTIL_GLOBAL': 'mean',
                          'etnia_si': 'sum','internet_si': 'sum',
                          'computador_si': 'sum','ID' : 'count',
                          'etnia_perc' : 'mean', 'int_perc' : 'mean', 'comp_perc': 'mean',
                          'cobertura_neta_secundaria' : 'mean',
                          'departamento': 'first',
                         'desercion_secundaria' : 'mean'}))
```

```python
departamento = departamento.astype({"code_dept":"int"})
```

```python
departamento.head()
```

```python
departamento.rename(columns= {'code_dept': 'code'}, inplace=True)
```

```python
departamento.reset_index()
```

```python
departamento.shape
```

```python
plt.figure(figsize=(6,6)) 
sns.scatterplot(data=departamento, x="PERCENTIL_GLOBAL", y="desercion_secundaria", hue = 'etnia_perc').set(title='Deserción y Rendimiento Escolar')
plt.xlabel('Percentil del Puntaje Global')
plt.ylabel('Deserción Secundaria')
plt.legend(title = '% Etnia')
plt.savefig('dept_percentil_desercion.png')
plt.show()
```
